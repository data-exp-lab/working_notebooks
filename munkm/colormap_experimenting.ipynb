```python
import numpy as np
import yt_pycanvas
import ipydatawidgets
from IPython.display import DisplayObject, display
import yt
```

We can access the colormaps widget within the FRB, so let's load an image through that to start. 

```python
ds = yt.load("/Users/madicken/Documents/yt_data/IsolatedGalaxy/galaxy0030/galaxy0030")
s = ds.r[:,:,0.5]
frb = yt_pycanvas.FRBViewer(height = 512, width = 512,
                            px = s["px"], py = s["py"], pdx = s["pdx"], pdy = s["pdy"],
                            val = s["density"])
display(frb)
```

To mainain compatibility for now, the frb can use the coloramps module but we can modify the colormaps module without affecting the frb. 

First, modify the f64 data array in the colormaps module. We should see this reflected in the js debugger and also a u8 array returned in the js debugger. 

```python
frb.colormaps.data = np.array([1.02, 2.01, 4.2], dtype=np.float64)
```

We can access the data array from the python side and it matches what we defined and what's in the js debugger. 

```python
frb.colormaps.data
```

but the image array is still empty, even though we can see them in the debugger on the js side. 

```python
frb.colormaps.image_array
```

We can also update the colormap name. (it won't update the frb yet) but the image array will be updated on the js side. 

```python
frb.colormaps.map_name = 'bone'
```
